```python
import pandas as pd
import json

i = 0
ein = []
file_num = []

while i < 8326:
    with open("/Users/mattkaplan/Desktop/all 990 JSON/allDAFs{}.json".format(i)) as f:
        data = json.load(f)
   
    ein.append(data["Return"]["ReturnHeader"]["Filer"]["EIN"])
    file_num.append(i)        
    print(i)
    i = i + 1
    
out_df = pd.DataFrame(columns=["EIN","file"])
out_df["EIN"]=ein
out_df["file_num"]=file_num

dups = out_df[out_df.EIN.duplicated(keep=False)]
dups.groupby("EIN")["file_num"].apply(list)
```
