```python
import xmltodict
import json

i = 0
while i < 8326:
    with open("/Users/mattkaplan/Desktop/Carnegie Mellon Research 2020/individual 990s (.XML)/allDAFs{}.xml".format(i)) as xml_file:
        data_dict = xmltodict.parse(xml_file.read())
    xml_file.close()
    
    json_data = json.dumps(data_dict)
    with open("allDAFs{}.json".format(i), "w") as json_file:
        json_file.write(json_data)
    json_file.close()
    i = i + 1
    print(i)
```
