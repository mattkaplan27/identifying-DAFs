```python
import pandas as pd
import requests
df_dafs = pd.read_csv("/Users/mattkaplan/Desktop/Thomas Ruchti/allDAFs.csv")

i = 0
while i < 100:
    url = df_dafs.iloc[i,0]
    response = requests.get(url)
    with open('sample_100{}.xml'.format(i), 'wb') as file:
        file.write(response.content)
    i = i+1
              
```
